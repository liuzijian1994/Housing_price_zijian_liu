```python
full_data = pd.concat((train, test)).reset_index(drop=True)
full_data.drop(['SalePrice'], axis=1, inplace=True)
null_count = full_data.isnull().sum().sort_values(ascending=False)
null_percentage = null_count / len(full_data)
null_data = pd.concat([null_count,null_percentage],axis=1,keys = ['count','percentage'])
print(null_data.head(30))
lenfull = len(full_data)
print(lenfull)
#check for nulls.
```

```python
# fill nulls for all data.
full_data["PoolQC"] = full_data["PoolQC"].fillna("None")
#NA means No Pool
full_data["MiscFeature"] = full_data["MiscFeature"].fillna("None")
# na means no
full_data["Alley"] = full_data["Alley"].fillna("None")
#na means no allay
full_data["Fence"] = full_data["Fence"].fillna("None")
#na means no fence
full_data["FireplaceQu"] = full_data["FireplaceQu"].fillna("None")
#na means no fireplace
full_data['LotFrontage']=full_data.groupby(['Neighborhood'])['LotFrontage'].transform(lambda row: row.fillna(row.median()))
#neighborhood are highly correlated with lotfrontage. besides, i don't have better alternative.
for column in ('GarageType', 'GarageFinish', 'GarageQual', 'GarageCond','BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2','MasVnrType','MSSubClass'):
    full_data[column] = full_data[column].fillna('None')
for column in ('GarageYrBlt', 'GarageArea', 'GarageCars','BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath','MasVnrArea'):
    full_data[column] = full_data[column].fillna(0)
null_count2 = full_data.isnull().sum().sort_values(ascending=False)
null_percentage2 = null_count2 / len(full_data)
null_data2 = pd.concat([null_count2,null_percentage2],axis=1,keys = ['count','percentage'])
null_data2.head(30)
# fill those nulls with actual meaning.
```

```python
full_data['MSZoning'] = full_data['MSZoning'].fillna(full_data['MSZoning'].mode()[0])
# mode() returns a DF, so i just need the first(biggest) number.
for column in ('Functional', 'Electrical', 'KitchenQual', 'Exterior1st','Exterior2nd','SaleType','Utilities'):
    full_data[column] = full_data[column].fillna(full_data[column].mode()[0])
null_countfinal = full_data.isnull().sum().sort_values(ascending=False)
null_percentagefinal = null_countfinal / len(full_data)
null_datafinal = pd.concat([null_countfinal,null_percentagefinal],axis=1,keys = ['count','percentage'])
null_datafinal.head(30)
# mode may not the best solution to fill actual nulls, but i think it`s enough.
# all nulls are gone.
```

```python
pd.set_option('display.max_rows',100)
print(full_data.dtypes.head(100))
pd.set_option('display.max_rows',20)
#accroding to varibale description, some numerical varibale are actually catagorical.
```

```python
for column in ('MSSubClass', 'MoSold', 'OverallCond', 'OverallQual','YrSold',):
    full_data[column] = full_data[column].astype(str)
print(format(full_data.shape))
num_columns = full_data.dtypes[full_data.dtypes != 'object'].index
full_data[num_columns] = np.log1p(full_data[num_columns])
#log transformation
```

```python
full_data = pd.get_dummies(full_data,drop_first=True)
print(format(full_data.shape))
train = full_data[:lentrain]
test = full_data[lentrain:]
print(format(train.shape))
print(format(test.shape))
#after get_dummy. num of variables increase from 79 to 300.
#split again
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
import xgboost as xgb
from sklearn.linear_model import Ridge, LassoCV
from sklearn.model_selection import cross_val_score
#Validation function
def rmse_cv(model):
    rmse= np.sqrt(-cross_val_score(model, train.values, train_y, scoring="neg_mean_squared_error", cv = 5))
    return(rmse)
```
